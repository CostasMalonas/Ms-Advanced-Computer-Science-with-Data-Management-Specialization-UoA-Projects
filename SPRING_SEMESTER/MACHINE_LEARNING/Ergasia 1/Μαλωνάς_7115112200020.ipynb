<h4>Άρα</h4>

\begin{document}

\begin{align*}
\frac{\partial}{\partial x} f(x) &= \frac{\partial}{\partial x} ( \mathbf{x}^T \mathbf{A} \mathbf{x} + \mathbf{b}^T \mathbf{x} ) \\
&= \frac{\partial}{\partial x} ( \mathbf{x}^T \mathbf{A} \mathbf{x} ) + \frac{\partial}{\partial x} ( \mathbf{b}^T \mathbf{x} ) \\
&= ( \mathbf{A} + \mathbf{A}^T ) \mathbf{x} + \mathbf{b} = \mathbf{Ax} + \mathbf{{A}^Tx} + \mathbf{b}
\end{align*}

\end{document}

\begin{document}
\[
\left|\left|A - XB\right|\right|_{F}^{2} = \mathrm{Tr}\left((A-XB)(A-XB)^{\mathrm{T}}\right) = \mathrm{Tr}\left((A-XB)(A^{\mathrm{T}}-X^{\mathrm{T}}B^{\mathrm{T}})\right) \Rightarrow \\
\left|\left|A - XB\right|\right|_{F}^{2} = \mathrm{Tr}\left(AA^{\mathrm{T}} - XBA^{\mathrm{T}} - AX^{\mathrm{T}}B^{\mathrm{T}} + XBB^{\mathrm{T}}X^{\mathrm{T}}\right) = \mathrm{Tr}\left(AA^{\mathrm{T}} - A^{\mathrm{T}}BX - AB^{\mathrm{T}}X^{\mathrm{T}} + XBB^{\mathrm{T}}X^{\mathrm{T}}\right) \Rightarrow \\
\left|\left|A - XB\right|\right|_{F}^{2} = \mathrm{Tr}\left(AA^{\mathrm{T}}\right) - \mathrm{Tr}\left(A^{\mathrm{T}}BX\right) - \mathrm{Tr}\left(\left(A^{\mathrm{T}}BX\right)^{\mathrm{T}}\right) +\mathrm{Tr}\left(XBB^{\mathrm{T}}X^{\mathrm{T}}\right) \\

\frac{\partial \left|\left|A - XB\right|\right|_{F}^{2}}{\partial X} = 0 - A^{\mathrm{T}}B - A^{\mathrm{T}}B + BB^{\mathrm{T}}X + B^{\mathrm{T}}BX = - 2A^{\mathrm{T}}B + BB^{\mathrm{T}}X + \left(BB^{\mathrm{T}}\right)^{\mathrm{T}}X = -2A^{\mathrm{T}}B + 2BB^{\mathrm{T}}X

<h3>Ερώτημα 3: Gradient descent</h3>

<h4>Ορίζουμε τις συναρτήσεις f1 και f2</h4>

```python
def f1(x1, x2):
    return (x1 - 2)**2 + (x2 - 3)**2

def f2(x1, x2):
    return (1 - (x2 - 3))**2 + 20*((x1 + 3) - (x2 - 3)**2)**2
```

<h4>Δημιουργία 2 ακολουθιών με αριθμούς από το -15 έως το 15 και βήμα 0.25</h4>

```python
x1 = np.arange(-15, 15, 0.25)
x2 = np.arange(-15, 15, 0.25)
#print(f1(x1, x2))
#print(f2(x1, x2))
```

<ul>
<li><b>df1dx1: </b>Επιστρέφει την μερική παράγωγο της f1 ως προς x1</li>
<li><b>df1dx2: </b>Επιστρέφει την μερική παράγωγο της f1 ως προς x2</li>
<li><b>df2dx1: </b>Επιστρέφει την μερική παράγωγο της f2 ως προς x1</li>
<li><b>df2dx2: </b>Επιστρέφει την μερική παράγωγο της f2 ως προς x2</li>
<li><b>grad_f1: </b>Επιστρέφει την κλίση της f1</li>
<li><b>grad_f2: </b>Επιστρέφει την κλίση της f2</li>
</ul>

```python
def df1dx1(x1):
    return 2*(x1 - 2)

def df1dx2(x2):
    return 2*(x2 - 3)

def df2dx1(x1, x2):
    return 40*(x1 - (x2 - 3)**2 + 3)

def df2dx2(x1, x2):
    return 2*(x2 - 2 - 40*(x1 - (x2 - 3)**2 + 3)*(x2 - 3))

def grad_f1(x1, x2):
    return np.array([df1dx1(x1), df1dx2(x2)])

def grad_f2(x1, x2):
    return np.array([df2dx1(x1, x2), df2dx2(x1, x2)])
```

```python
x = 0
y = 5
h = 0.001
print((f1(x+h, y)-f1(x, y))/h)
print(df1dx1(x))
print((f1(x, y+h)-f1(x, y))/h)
print(df1dx2(y))
print(grad_f1(x, y))
```

<h3>Ορίζουμε την συνάρτηση grad_descent για την εκτέλεση του gradient descent</h3>

```python
def grad_descent(f, grad_f, init_t, alpha, max_iter):
    eps = 1e-5
    prev_t = init_t-10*eps
    t = init_t.copy()
    iter_num = 0
    value_at_end_of_iteration = []
    while norm(t - prev_t) > eps and iter_num < max_iter:
        prev_t = t.copy()
        t -= alpha*grad_f(t[0], t[1])
        print(f"Επανάληψη {iter_num}, f(x) = {f(t[0], t[1])}, κλίση = {grad_f(t[0], t[1])}")
        value_at_end_of_iteration.append(f(t[0], t[1]))
        iter_num += 1

    return t, value_at_end_of_iteration
```

```python
print(f1(0,0))
print(f2(0,0))
```

<h3>Συνάρτηση plot_graph για να κάνουμε plot τα αποτελέσματα μας για εξαγωγή συμπερασμάτων</h3>

```python
def plot_graph(values_at_each_iter):
    iterations = list(range(len(values_at_each_iter)))
    plt.plot(iterations, values_at_each_iter)
    plt.show()
```

<h3>Δοκιμές για την συνάρτηση f1</h3>

```python
x, y = grad_descent(f1, grad_f1, np.array([0.0, 0.0]), 0.5, 10)
plot_graph(y)
```

Παρατηρούμε πως για learning rate ίσο με 0.5 και για 10 επαναλήψεις η συνάρτηση f1 συγκλίνει κατευθείαν στο 0.

Για μικρότερο learning rate η συνάρτηση μας συγκλίνει μετά από τις 10 επαναλήψεις
